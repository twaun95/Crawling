```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pymysql
import time
```

```python
def chosunnews(search):
    url = "http://www.newsis.com/search/schlist/?val={search}&sort=acc&jo=all_jogun&bun=all_bun&sdate=&term=allday&edate=&s_yn=Y&catg=1&t=1&".format(search=search)
    driver = webdriver.Chrome("./chromedriver.exe")
    driver.get(url)

    chosunnewsUrlList(driver)
```

```python
def chosunnewsUrlList(driver):
    url2 = driver.current_url
    
    for pageNo in range(1,3,1): #2페이지까지
        print("1")
        url = url2 + "&page={pageNo}".format(pageNo=pageNo)
        driver.get(url)
        time.sleep(2)
        print("2")
        chosunnewsUrs= driver.find_elements_by_css_selector(".lst_p12 .group .area .title a")
        chosunnewsUrs_List = []        
        for urlList in chosunnewsUrs:
            chosunnewsUrs_List.append(urlList.get_attribute("href"))

        chosunnewsInfo(driver,chosunnewsUrs_List)
```

```python
def chosunnewsInfo(driver,chosunnewsUrs_List):
    for urlList in chosunnewsUrs_List:
        driver.get(urlList)
        time.sleep(2)
        print("3")

        newsUrl = urlList
        newsTitle = driver.find_element_by_css_selector("#article .article_tbx h1").text
        print("4")
        newsSubtitles = driver.find_elements_by_css_selector("#textBody .summary_view")
        newSubtitle_result = ""
        for newsSubtitle in newsSubtitles:
            newSubtitle_result+=newsSubtitle.text
            
        newsDates = driver.find_elements_by_css_selector(".function .rgroup .date")
        newsDate_result = ""
        for newsDate in newsDates:
            newsDate_result+=newsDate.text
        
        newsDate_result = newsDate_result.replace("등록 ","").replace("-","")[0:8]
            
        newsContents = driver.find_elements_by_css_selector("#textBody")
        content_result = ""
        for content in newsContents:
            content_result+=content.text

        print("- 기사 URL - : \n",newsUrl)
            
        print("- 기사 제목 - : \n",newsTitle)
        print("- 기사 부제목 - \n:",newSubtitle_result)
        print("- 기사 날짜 - \n:",newsDate_result)
        
        print("- 기사 본문 - \n")


        print(content_result)
            
        dbData = [[newsUrl, newsTitle, newsDate_result, newSubtitle_result,content_result]]

#         connectDB(dbData)
```

```python
def connectDB(dbData):
    DB_HOST = '127.0.0.1'
    DB_USER = 'root'
    DB_PASSWD = 'autoset'
    DB_NAME = 'python'
    print("5")

    conn = pymysql.connect(host=DB_HOST, user=DB_USER, password=DB_PASSWD,
                       db=DB_NAME, charset='utf8')
    
    curs = conn.cursor()

    sql = """insert into chosunnews(newsUrl, newsTitle, newsDate, newSubtitle, content)
         values (%s, %s, %s, %s, %s)"""
    curs.executemany(sql,dbData)
    
    conn.commit()

    conn.close()
    
```
